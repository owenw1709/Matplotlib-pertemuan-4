Import Module

```python
%matplotlib inline
```

```python
import matplotlib
import matplotlib.pyplot as plt

print (matplotlib.__version__)
```

## Simple Pie Plot

```python
kategori =['A','B','C','D','E']
data = [215,130,245,210,225]
```

```python
plt.pie(data,
       labels=kategori,
       autopct='%1.1f%%',
        startangle=90)

plt.title ('Contoh Pie Plot')

plt.show()
```

## Pengaturan Legends dan Color 

```python
kategori =['A','B','C','D','E']
data = [215,130,245,210,225]
warna = ['pink','cyan','#E03364','yellowgreen','skyblue']
```

```python
plt.pie(data,
       labels=kategori,
        colors= warna,
       autopct='%1.1f%%',
        startangle=90)
plt.legend()
plt.title ('Contoh Pie Plot')

plt.show()
```

## Exploded Pie Plot 

```python
kategori =['A','B','C','D','E']
data = [215,130,245,210,225]
warna = ['pink','cyan','#E03364','yellowgreen','skyblue']
explode_var = [0.,0.,0.2,0.,0.08]
```

```python
plt.pie(data,
       labels=kategori,
        colors= warna,
        explode=explode_var,
       autopct='%1.1f%%',
        startangle=90)
plt.legend()
plt.title ('Contoh Exploded Pie Plot')

plt.show()
```
